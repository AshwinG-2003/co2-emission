## Introduction
Carbon dioxide emissions are a primary factor of global climate change. It’s widely recognised that to avoid the worst impacts of climate change, the world needs to urgently reduce emissions. I have decided to find a dataset that shows the carbon dioxide emissions emitted from various countries. This data was found and imported from https://www.kaggle.com/datasets/koustavghosh149/co2-emission-around-the-world. This dataset includes numerical data for carbon dioxide emissions emitted by 215 countries from 1990 to 2019. Throughout this dataset I will first explore the number of countries that emit carbon dioxide within each region. Along with that I will analyze the region and country that emits the most amount of carbon dioxide. Finally I will explore the amount of carbon dioxide emitted from 1990 to 2019 from all countries. Throughout my analysis I will see the relationship between the amount of countries and the regions that emit carbon dioxide. This analysis could show which countries emit the most amount of carbon dioxide and show how carbon emissions have been gradually increasing or decreasing from 1990 to 2019.

## Methods
### Data Collection


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing data from CSV File

```python
emission = pd.read_csv("CO2_emission.csv")
```

### Data Cleaning

This dataset had 31 countries with missing data for the amount of carbon dioxide emission so I have removed them. Along with that I also have removed the country code, indicator name and 2019.1 from the columns.

```python
emission = emission.dropna(axis=0)
emission = emission.drop(["country_code", "Indicator Name", "2019.1"], axis=1)
emission
```

## Analyses
I will explore the relationship between the number of countries that emit carbon dioxide within each region and the amount of carbon dioxide emitted by each region. Along with that I will find the country that emits the most amount of CO2 emissions.


### Distributions


I will begin to analyze the amount of countries that emits CO2 within each region. The figure below shows a visual representation of the regions that emit the most amount of CO2. From this we can see that Europe and Central Asia had the most amount of countries that emit CO2 while North America had the least amount of countries that emit CO2.

```python
countCountry = emission["Country Name"].groupby([emission["Region"]])
countCountry.count()
```

```python
sns.set(rc = {'figure.figsize':(20,8)})
fig1 = sns.countplot(x = 'Region', data = emission, palette ='Set1')
fig1 = fig1.set(title = "Number of Countries that Emit CO2 in Each Region", xlabel = 'Region', ylabel = 'Number of Countries')
```

### Region that Emitted the Most Amount of CO2
Now I will be looking at the amount of CO2 emitted from each region over the timespan between 1990 to 2019.

```python
mean = emission.groupby("Region").mean().mean(axis=1)
mean
```

The figure below shows a visual representation of amount of CO2 emitted by each region. As we can see that North America has the most amount of CO2 emitted of 17.036275 ton and South Asia has the least amount of CO2 emitted of 0.726849 ton. From this analysis we can see that the number of countries does not affect the amount of CO2 emitted as there were only 2 countries that emit CO2 from North America and surprisingly they emit the most amount of CO2.

```python
fig2 = emission.groupby("Region").mean().mean(axis=1)
fig2.plot.bar(xlabel="Region", ylabel="CO2 Emissions(Ton)", title="Number of CO2 Emissions in Each Region")
```

### Country that Emitted the Most Amount of CO2
Now I will find the country that emits the most amount of CO2 within each region.

```python
max = emission.groupby("Region").max()
countryMax = max.groupby("Country Name").mean().mean(axis=1)
countryMax
```

The figure below shows a visual representation of the country that emitted the most amount of CO2 within each region. From this we can which countries emit the most amount of CO2 from each region.

```python
fig3 = max.groupby("Country Name").mean().mean(axis=1)
fig3.plot.bar(xlabel="Country", ylabel="CO2 Emissions(Ton)", title="Countries that Emitted the Most Amount of CO2")
```

### CO2 Emissions from 1990 to 2019
Finally I will be analyzing the CO2 emitted over the timespan between 1990 to 2019. 


I have first shown the amount of CO2 emitted each year from all the countries.

```python
meanCO2 = emission.groupby("Region").mean().mean(axis=0)
meanCO2
```

The figure below shows a line graph of the amount of CO2 emitted from 1990 to 2019. From this we can see that it has been gradually increasing over the years until 2007 and then it started to decrease as more electric vehicles have been introduced and people are becoming more eco friendly.

```python
fig4 = emission.groupby("Region").mean().mean(axis=0)
fig4.plot(xlabel="Year", ylabel="CO2 Emissions(Ton)", title="Amount of CO2 Emissions from 1990 to 2019")
```
